```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def load_time_series(filepath):
    """Load time series data from a CSV file"""
    df = pd.read_csv(filepath, parse_dates=['Year'])
    df.set_index('Year', inplace=True)
    return df

def plot_time_series(df):
    """Plot time series data for different sources"""
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=df, x=df.index, y='Mean', hue='Source', marker='o')
    plt.xlabel("Year")
    plt.ylabel("Mean Value")
    plt.title("Time Series Data Visualization")
    plt.legend(title='Source')
    plt.grid()
    plt.show()

def plot_moving_average(df, window=5):
    """Plot time series with moving average smoothing"""
    df_ma = df.copy()
    df_ma['Smoothed'] = df_ma['Mean'].rolling(window=window, min_periods=1).mean()
    
    plt.figure(figsize=(12, 6))
    plt.plot(df_ma.index, df_ma['Mean'], label='Original', alpha=0.5)
    plt.plot(df_ma.index, df_ma['Smoothed'], label=f'{window}-Period Moving Average', color='red')
    plt.xlabel("Year")
    plt.ylabel("Mean Value")
    plt.title("Moving Average Smoothing")
    plt.legend()
    plt.grid()
    plt.show()

def main():
    filepath = "C://Users//jayas//OneDrive//Documents//VScode//TimeSeriesAnalysis//globaltemp.csv"
    df = load_time_series(filepath)
    plot_time_series(df)
    plot_moving_average(df)
    
if __name__ == "__main__":
    main()
```
